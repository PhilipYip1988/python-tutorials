### series and dataframe recap

## figure, axes and plot

We should conceptualise a graph as a figure, which can be thought of as a blank canvas which can contain one or more axes and an axes in turn as a container which can contain different plot types.

To create a new figure, we can explicitly use the pyplot plotting function ```figure```.

````python
plt.figure()
````

This creates an instance of the ```Figure``` class also known as a ```Figure``` object. 

In the cell output above, we do not see this figure because it has no axes.

\<Figure size 432x288 with 0 Axes\> is displayed twice:

* The first occurance displays at the top of the cell output and highlights a ```Figure``` object is created, that is essentially a blank canvas. 
* The second occurance displays because the ```Figure``` object is not assigned to a variable name so details about the last matplotlib object created are displayed at the bottom of the cell output. This can be suppressed in the cell output by appending the function call with ```;```

````python
plt.figure();
````

To create an ```AxesSubplot``` object on the ```Figure``` object we can explicitly use the pyplot function ```axes```:

We create a ```Figure``` object and then add a single ```AxesSubplot``` object to it using the pyplot function ```axes```:

````python
plt.figure()
plt.axes()
````

Note the ```AxesSubplot``` object will be added to the currently selected ```Figure``` object and by default span to fit the canvas of the ```Figure``` object. The ```AxesSubplot``` object has an x axis and a y axis. The initial x scale and y scale are between 0.0 and 1.0 and by default these values will automatically scale to the data when plots are added to the ```AxesSubplot``` object.

If the pyplot function ```axes``` is called without previously creating a ```Figure``` object, then a new ```Figure``` object will be created and be currently selected:

````python
plt.axes()
````

The pyplot function ```plot``` can be used to add a line plot. The line plot adds a list of ```Line2D``` objects to the currently selected ```AxesSubplot```. In the simplist case this is a list of a single ```Line2D``` object. 

````python
plt.figure()
plt.axes()
plt.plot(xdata, ydata)
````

In ```[<matplotlib.lines.Line2D>\]```, the ```[ ]``` indicates a list and this list has a single ```Line2D``` object. We can suppress these additional details by ending the function calls with the ```;```

````python
plt.figure();
plt.axes();
plt.plot(xdata, ydata);
````

If the pyplot function ```plot``` is called without creating a ```Figure``` object, or ```AxesSubplot``` object then a new ```Figure``` and ```AxesSubplot``` will automatically be created and selected:

````python
plt.plot(xdata, ydata);
````

### keyword input arguments

In the above case we used the pyplot plotting functions ```figure``` and ```axes``` without specifying any keyword input arguments and therefore used their default values.

Input ```plt.figure(``` and then press ```⇧``` and ```↹```:

Input ```plt.axes(``` and then press ```⇧``` and ```↹```:

````python
? plt.figure
````

The ```figure``` function for example has the keyword input arguments ```num```, ```figsize``` and ```dpi```. 

```num``` by default is configured to use the next integer. The first figure will have ```num=1```, the second will have ```num=2``` and so on. 

````python
plt.figure(num=1);
plt.plot(xdata, ydata);
plt.plot(xdata, 2*ydata);
plt.plot(xdata, 3*ydata);

plt.figure(num=2);
plt.plot(xdata, 2*ydata);

plt.figure(num=3);
plt.plot(xdata, 3*ydata);
````

If ```num``` is assigned to the ```num``` of an existing ```figure``` it will be reselected:

````python
plt.figure(num=1);
plt.plot(xdata, ydata);

plt.figure(num=2);
plt.plot(xdata, 2*ydata);

plt.figure(num=2);
plt.plot(xdata, 3*ydata);
````

The ```figsize``` keyword input argument is a tuple of ```(width, height)``` and is used to specify the dimensions of the ```Figure``` object canvas.

````python
plt.figure(figsize=(4,3));
plt.plot(xdata, ydata);

plt.figure(figsize=(8,6));
plt.plot(xdata, ydata);

plt.figure(figsize=(1, 5));
plt.plot(xdata, ydata);
````

The dots per inch ```dpi``` keyword input argument can be used to increase or decrease the resolution of the figure and has a default value of 100:

````python
plt.figure(figsize=(4,3), dpi=100);
plt.plot(xdata, ydata);

plt.figure(figsize=(4,3), dpi=200);
plt.plot(xdata, ydata);

plt.figure(figsize=(8,6));
plt.plot(xdata, ydata);
````

The pyplot ```axes``` function has an optional positional input arguments ```rect``` which is in the form of a tuple (or can be supplied as a list). The tuple must contain four values corresponding to the dimensions of a rectangle ```(left, bottom, width, height)```. These values use the co-ordinate system of the figure which is normalised between ```0.0``` and ```1.0```.

The bottom left of the figure will therefore have a position of ```left=0.0``` and ```bottom=0.0``` and the entire length of the figure is ```width=1.0``` and ```height=0.0```. We can compare the defaults which has approximately ```(0.0, 0.0, 0.78, 0.78)``` to ```(0.0, 0.0, 1.0, 1.0)```:

````python
plt.figure();
plt.axes();
plt.figure();
plt.axes((0.0, 0.0, 0.78, 0.78));
plt.figure();
plt.axes((0.0, 0.0, 1.0, 1.0));
````

We can also create multiple ```AxesSubplot``` objects on the ```Figure``` object:

````python
plt.figure();
plt.axes((0.0, 0.0, 1.0, 1.0));
plt.axes((0.55, 0.55, 0.4, 0.4));
````

````python
plt.figure();
plt.axes((0.0, 0.0, 0.4, 0.4));
plt.axes((0.1, 0.1, 0.4, 0.4));
plt.axes((0.2, 0.2, 0.4, 0.4));
plt.axes((0.3, 0.3, 0.4, 0.4));
````

````python
plt.figure();
plt.axes((0.0, 0.0, 0.4, 0.4));
plt.plot(xdata, ydata);
plt.axes((0.1, 0.1, 0.4, 0.4));
plt.plot(xdata, 2*ydata);
plt.axes((0.2, 0.2, 0.4, 0.4));
plt.plot(xdata, 3*ydata);
plt.axes((0.3, 0.3, 0.4, 0.4));
plt.plot(xdata, 4*ydata);
````

````python
? plt.legend
````

## tight_layout

Sometimes when adding a long string to the xlabel, ylabel, title and legend, the text can begin to run into each other. This is more prevalent when multiple AxesSubplots are used. The pyplot function ```tight_layout``` attempts to address these issues:

````python
plt.figure();
plt.axes();
plt.plot(xdata, ydata, label="y=sin(x)");
plt.plot(xdata, 2*ydata, label="y=2sin(x)");
plt.plot(xdata, 3*ydata, label="y=3sin(x)");
plt.xlabel("x");
plt.ylabel("y");
plt.title("y=sin(x)");
plt.legend(bbox_to_anchor=(1.05, 0.6));
````

````python
plt.figure();
plt.axes();
plt.plot(xdata, ydata, label="y=sin(x)");
plt.plot(xdata, 2*ydata, label="y=2sin(x)");
plt.plot(xdata, 3*ydata, label="y=3sin(x)");
plt.xlabel("x");
plt.ylabel("y");
plt.title("y=sin(x)");
plt.legend(bbox_to_anchor=(1.05, 0.6));
plt.tight_layout();
````

## xscale, xlim, yscale, ylim

Let us now have a look at a new array:

````python
xdata = np.linspace(start=0, stop=10, num=11)
xdata
````

````python
ydata = np.exp(xdata)
ydata
````

Here we see for small numbers of x, the corresponding value of y is very small however for larger values of x, the corresponding value of y is very large:

````python
pd.DataFrame({"x": xdata,
              "y": ydata})
````

This means for a standard plot we cannot really see what is happening at the lower values:

````python
plt.figure();
plt.axes();
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();
````

We can use the pyplot functions ```xlim``` to set the ```left``` and ```right``` limits or the ```ylim``` to set the ```lower``` and ```upper``` limits on the x and y axis respectively. Input ```plt.xlim(``` and then press ```⇧``` and ```↹```:

````python
? plt.xlim
````

Input ```plt.ylim(``` and then press ```⇧``` and ```↹```:

````python
? plt.ylim
````

If for example we want to see the lower regime, we can use the keyword inputs ```left``` and ```right``` for the pyplot function ```xlim``` and ```bottom``` and ```top``` for the pyplot function ```ylim```:

````python
plt.figure();
plt.axes();
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xlim(left=0, right=4);
plt.ylim(bottom=-1, top=50);
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();
````

We can also input these limits as a tuple:

````python
plt.figure();
plt.axes();
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xlim((0, 4));
plt.ylim((-1, 50));
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();
````

We can use the pyplot function ```axes``` to create an ```AxesSubplot``` to display the standard plot and nest another ```AxesSubplot``` to examine in more detail smaller regimes of x and y:

````python
plt.figure();

plt.axes((0.0, 0.0, 1.0, 1.0));
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();

plt.axes((0.10, 0.45, 0.4, 0.4));
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xlim((0, 4));
plt.ylim((-1, 50));
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();
````

So far we have only used a linear scale. The logarithmic scale can instead be used which emphasises changes in smaller values. We can change the scale of each axis using the pyplot functions ```xscale``` and ```yscale``` respectively. Input ```plt.xscale(``` and then press ```⇧``` and ```↹```:

````python
? plt.xscale
````

Input ```plt.ylim(``` and then press ```⇧``` and ```↹```:

````python
? plt.yscale
````

Let us explicitly use the pyplot function ```xscale``` to set the ```value``` to the string ```"linear"``` (the default value) and ```yscale``` to set the ```value``` to the string ```"log"```:

````python
plt.figure();
plt.axes();
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="log");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="log(y)");
plt.grid();
````

The numpy ```log``` and ```exp``` functions are inverse functions and therefore we see a straight line.

````python
plt.figure();

plt.axes((0.0, 0.0, 0.5, 1.0));
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="linear");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();

plt.axes((0.5, 0.0, 0.5, 1.0));
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="log");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="log(y)");
plt.grid();
````

The function has a variable number of positional input arguments. Normally we use three which have the form nrows, ncols and index. For simplicity (although this isn't commonly done) we will use this function to create a single subplot:

And use these to plot the linear and log plot of the exp function:

````python
# figure
plt.figure();

# left subplot
plt.subplot(1, 2, 1);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="linear");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();

# right subplot
plt.subplot(1, 2, 2);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="log");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="log(y)");
plt.grid();
````

````python
# figure
plt.figure(figsize=(15,5));

# left subplot
plt.subplot(1, 2, 1);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="linear");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();

# right subplot
plt.subplot(1, 2, 2);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="log");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="log(y)");
plt.grid();
````

Using the pyplot ```tight_layout``` function can also help:

````python
# figure
plt.figure(figsize=(15,5));

# left subplot
plt.subplot(1, 2, 1);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="linear");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="y");
plt.grid();

# right subplot
plt.subplot(1, 2, 2);
plt.minorticks_on();
plt.plot(xdata, ydata);
plt.xscale(value="linear");
plt.yscale(value="log");
plt.xlabel(xlabel="x");
plt.ylabel(ylabel="log(y)");
plt.grid();

# tight layout
plt.tight_layout();
````
