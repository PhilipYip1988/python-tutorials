# IPython Magics (Linux)

## Shells

In order to prevent confusion between programming languages, the following recap will be made.

The Linux Terminal uses bash by default.

bash Prompt:

```bash
(base) username@pc:~$
```

bash can be used to launch Python:

```bash
(base) username@pc:~$ python
```

Python Prompt:

```python
>>> 
```

The Python program can be exited using the ```exit``` function returning to the bash Prompt:

```python
>>> exit()
(base) username@pc:~$
```

bash can be used to launch IPython:

```bash
(base) username@pc:~$ ipython
```

IPython Prompt:

```python
[1] 
```

The IPython program can be exited using the ```exit``` function returning to the bash Prompt:

```python
[1] exit
(base) username@pc:~$
```

## IPython Cells

An interactive Python notebook uses an IPython shell by default. This is why all the cells are numbered like in IPython when executed:

````python
print('Hello World!')
````

## IPython Magics

IPython also has access to IPython Magics. These are reimplementations of bash commands in Python. 

```%lsmagic``` can be used to list the IPython magics. These are split into line and cell magics. In a notebook the magics will output using JSON and one can interactively examine the output:

````python
%lsmagic
````

These can also be assigned to a variable and printed:

````python
magic_commands = %lsmagic
````

````python
print(magic_commands)
````

A quick reference of the commands can be seen by using ```%quickref```:

````python
%quickref
````

More details will show when the line magic ```%magic``` is used:

````python
%magic
````

A line magic is normally prefixed with a ```%``` which indicates an IPython magic is used:

````python
%ls
````

Because automagic is on, it is also possible to use a line magic without the ```%``` sign but less common to do so:

````python
%automagic?
````

````python
ls
````

Help about the line magic can be viewed using:

````python
%ls?
````

A cell magic begins with ```%%``` and occupies a cell:

The cell magic ```%%writefile``` is used to write a file:

````python
%%writefile?
````

And has the alias ```%%file```:

````python
%%file?
````

A Python script can be created using:

````python
%%writefile script.py
print('Hello World!')
print('Hello World!')
````

This script can be run using the line magic ```%run```:

````python
%run script.py
````

It can be loaded into a cell using the line magic ```%load```:

````python
# %load script.py
print('Hello World!')
print('Hello World!')
````

```%run``` is an IPython magic used to execute a Python script and ```%%python``` is a cell magic used to run Python code (opening a Python Shell within an IPython Cell):

````python
%%python
print('Hello World!')
print('Hello World!')
````

If the following variables are created:

````python
greeting = 'hello'
greeting_b = b'hello'
whole_num = 2
num = 3.14
boolean = True
archive = (1, 2, 3, 1)
active = [1, 2, 3, 1]
unique = {1, 2, 3, 4}
mapping = {'r': 'red', 'g': 'green', 'b': 'blue'}
````

Variables can be listed using:

````python
dir()
````

However this lists all the additional variables in the main namespace created by IPython. Instead ```%who_ls``` can be used to only list user created variables:

````python
%who_ls
````

The relation ipython magic ```%who``` will print user created variables separated by a tab:

````python
%who
````

More details can be seen using ```%whos```:

````python
%whos
````

Additionally, variables may be filtered using their ```type```:

````python
%whos str
````

## Bash Commands

A Bash command can be run directly using the prefix ```!```, the Bash command ```python``` that the line magic ```%run``` and cell magic ```%%python``` are based on can be used to run a script file. Because the Python shell closes after the script file is executed, this works as expected:

````python
!python script.py
````

If however the following is run:

```python
!python
print('Hello World!')
print('Hello World!')
```

The IPython cell will freeze as the Python program has not closed. Generally a preference should be made for IPython magics when available over their original Bash counterparts as they have been updated to work properly within an IPython cell:

Details about these two commands can be seen using ```?```

````python
%run?
````

````python
%%python?
````

The list of commonly used bash commands can be viewed using the bash command ```help```, recall this is prefixed with a ```!```:

````python
!help
````

If the script file is recreated:

````python
%%writefile script.py
print('Hello World!')
print('Hello World!')
````

The IPython line magic ```%alias``` can be used to view the list of alias, the ones shown are available by default:

````python
%alias
````

A new alias can be created using this line magic:

````python
%alias?
````

In bash the command ```touch``` can be used to touch a directory creating a new empty file:

````python
!touch untitled1.txt
````

An alias to touch can be made using:

````python
%alias touch touch
````

Now:

````python
%touch untitled2.txt
````

And because automagics are enabled:

````python
touch untitled3.txt
````

These three files can be removed using:

````python
rm untitled1.txt untitled2.txt untitled3.txt
````

The ```echo``` command is the bash equivalent for a ```print``` statement:

````python
!echo 'Hello World!'
````

It is not defined as an alias:

````python
%alias echo echo
````

````python
%echo 'Hello World!'
````

## Magics for File and Folders

The cell magic ```%%writefile``` can be used to create a new file:

````python
%%writefile script.py
print('Hello World!')
````

The cell magic ```%cp``` can be used to copy the file:

````python
%cp script.py script2.py
````

The line magics ```%mkdir``` and ```%rmdir``` can be used to make and remove directories. Note that use of a space will make separate folders:

````python
%mkdir new folder
````

The line magic ```%ls``` will list files and directories:

````python
%ls?
````

````python
%ls
````

The option to list 1 file per line can be selected:

````python
%ls -1
````

The directories can be removed using:

````python
%rmdir new
%rmdir folder
````

A single quotation can be used in a directories name:

````python
%mkdir 'new folder'
````

````python
%ls
````

````python
%rmdir 'new folder'
````

Double quotations can't be used in a directory name so are used to enclose names which include spaces:

````python
%mkdir "new folder"
````

````python
%ls
````

The current working directory can be printed:

````python
%pwd
````

The IPython reimplementation allows for this return value to be assigned to a variable:

````python
current_directory = %pwd
````

The line magic ```%cd``` can be used to change the directory:

````python
%cd?
````

When no input argument is made the directory will change to:

````python
%cd
````

Which is the same as using ```~```:

This can be changed to Documents using:

````python
%cd ~/Documents
````

To change back to the original current working directory, the variable needs to be inserted using braces:

````python
%cd {current_directory}
````

The line magic ```%bookmark``` can be used to bookmark a directory:

````python
%bookmark Documents ~/Documents
````

And this bookmarked directory can be used with ```%cd```:

````python
%cd Documents
````

Changing back to the working directory of the notebook file:

````python
%cd {current_directory}
````

A file can be renamed using the line magic move ```%mv```:

````python
%mv script2.py script3.py
````

````python
%ls -1
````

The cell magic ```%%script``` can be used to launch a script using the specified programming language:

````python
%%script?
````

Each supported programming language has an alias ```%%python``` (```%%python3```) and ```%%bash``` and ```%%perl```. 

Examples of ```%python``` have already been examined. ```%%sh``` and ```%%python2``` are older versions of the Linux and Python shells and are generally no longer used.

````python
%%bash?
````

````python
%%python?
````

The additional files and folders created in this notebook will be deleted, in order for it to be run again wirthout any errors:

````python
%rmdir "new folder"
%rm script.py script3.py
````

## Other Magics

```%pdef```, ```%pdoc``` are line magics used to print the definition of a function and the docstring of a function respectively. Normally these are combined in the line magic ```%pinfo```, some functions may also have extended information which can be accessed using the line magic ```%pinfo2```. Because the line magic ```%pinfo``` is so commonly used it is implemented using the alias ```?```

````python
print?
````

````python
%pdef print
````

````python
%pdoc print
````

````python
%pinfo print
````

````python
%pinfo2 print
````

The line magic ```%matplotlib``` is used to change the configuration options for ```matplotlib```:

````python
%matplotlib?
````

The backends can be viewed using:

````python
%matplotlib --list
````

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
````

A plot can be made using the default inline backend of a notebook:

````python
plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 10])
````

Or it can be plotted in an interactive window using qt:

````python
%matplotlib qt
````

````python
plt.plot([1, 2, 3, 4, 5], [2, 4, 6, 8, 10])
````

The line magic ```%pprint``` can be used to change pretty print settings which will attempt to pretty print collections such as ```list``` and ```dict``` instances:

````python
%pprint?
````

For example the ```dict``` instance ```CSS4_COLORS``` can be examined with and without pretty printing:

````python
%pprint
````

````python
mpl.colors.CSS4_COLORS
````

````python
%pprint
````

````python
mpl.colors.CSS4_COLORS
````

The line magic ```%history``` gives the command history:

````python
%history?
````

It has the alias ```%hist```:

````python
%hist?
````

````python
%history
````

The package manager can be accessed using:

````python
%conda
````

And the subcommand ```list``` can be used to list all the packages in the Python environment:

````python
%conda list
````

The line magics ```%conda```, ```%mamba``` and ```%pip``` are for the respective package managers. For a conda environment only the ```conda``` package manager should be used. ```conda``` now uses the ```libmamba``` solver by default.
