```python
import tkinter
from tkinter import Tk, Button, Label, Entry, END
tk = tkinter.Tk()
tk.title("Simple Calculator")

entryBox = Entry(tk, width=40, borderwidth=5)
entryBox.grid(row=0, column=0, columnspan=4,
             padx=10, pady=10)

def button_click(number):
    oldEntryText = entryBox.get()
    newEntryText = str(oldEntryText) + str(number)
    entryBox.delete(0, END)
    entryBox.insert(0, newEntryText)
    return None

def button_clear():
    entryBox.delete(0, END)
    return None

def button_operator(operator):
    oldEntryText = entryBox.get()
    newEntryText = str(oldEntryText) + operator
    entryBox.delete(0, END)
    entryBox.insert(0, newEntryText)
    return None

def button_equal():
    oldEntryText = entryBox.get()
    
    if oldEntryText.find('+') != -1:
        oldEntryTextList = oldEntryText.split("+")
        result = int(oldEntryTextList[0]) + int(oldEntryTextList[1])
    elif oldEntryText.find('-') != -1:
        oldEntryTextList = oldEntryText.split("-")
        result = int(oldEntryTextList[0]) - int(oldEntryTextList[1])      
    elif oldEntryText.find('*') != -1:
        oldEntryTextList = oldEntryText.split("*")
        result = int(oldEntryTextList[0]) * int(oldEntryTextList[1])         
    elif oldEntryText.find('/') != -1:
        oldEntryTextList = oldEntryText.split("/")
        result = int(oldEntryTextList[0]) // int(oldEntryTextList[1])          
    else:
        newEntryText = "Invalid Result"
        
    newEntryText = str(result)
    entryBox.delete(0, END)
    entryBox.insert(0, newEntryText)
    return None


# Create Buttons
buttonZero = Button(tk, text="0", padx=40, pady=20, 
                    command=lambda: button_click(0))
buttonOne = Button(tk, text="1", padx=40, pady=20, 
                   command=lambda: button_click(1))
buttonTwo = Button(tk, text="2", padx=40, pady=20, 
                   command=lambda: button_click(2))
buttonThree = Button(tk, text="3", padx=40, pady=20, 
                     command=lambda: button_click(3))
buttonFour = Button(tk, text="4", padx=40, pady=20, 
                    command=lambda: button_click(4))
buttonFive = Button(tk, text="5", padx=40, pady=20, 
                    command=lambda: button_click(5))
buttonSix = Button(tk, text="6", padx=40, pady=20, 
                   command=lambda: button_click(6))
buttonSeven = Button(tk, text="7", padx=40, pady=20, 
                     command=lambda: button_click(7))
buttonEight = Button(tk, text="8", padx=40, pady=20, 
                     command=lambda: button_click(8))
buttonNine = Button(tk, text="9", padx=40, pady=20, 
                    command=lambda: button_click(9))

buttonAdd = Button(tk, text="+", padx=40, pady=20, 
                   command=lambda: button_operator("+"))
buttonSubtract = Button(tk, text="-", padx=40, pady=20, 
                        command=lambda: button_operator("-"))
buttonMultiply = Button(tk, text="*", padx=40, pady=20, 
                        command=lambda: button_operator("*"))
buttonDivide = Button(tk, text="/", padx=40, pady=20, 
                      command=lambda: button_operator("/"))

buttonEqual = Button(tk, text="=", padx=86, pady=20, command=button_equal)
buttonClear = Button(tk, text="CLS", padx=80, pady=20, command=button_clear)

# Place the buttons on the screen:
buttonSeven.grid(row=1, column=0)
buttonEight.grid(row=1, column=1)
buttonNine.grid(row=1, column=2)
buttonFour.grid(row=2, column=0)
buttonFive.grid(row=2, column=1)
buttonSix.grid(row=2, column=2)
buttonOne.grid(row=3, column=0)
buttonTwo.grid(row=3, column=1)
buttonThree.grid(row=3, column=2)
buttonZero.grid(row=4, column=0)

buttonClear.grid(row=5, column=0, columnspan=2)
buttonEqual.grid(row=5, column=2, columnspan=2)

buttonAdd.grid(row=1, column=3)
buttonSubtract.grid(row=2, column=3)
buttonMultiply.grid(row=3, column=3)
buttonDivide.grid(row=4, column=3)



tk.mainloop()
```
