# Plotting Methods

## Importing the Data Science Libraries

````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
````

## Line2D

The Axes plotting function plot creates a list of Line2D plots. For linear data this looks like:

````python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0.001, 2.001, 3.999, 5.999, 8.002, 10.001])
````

````python
fig, ax = plt.subplots()
ax.plot(x, y);
````

The Axes plotting function steps is equivalent to the Axes plot method with a drawstyle preconfigured to 'steps_pre':

````python
fig, ax = plt.subplots(nrows=1, ncols=2);
ax[0].plot(x, y, drawstyle='steps-pre');
ax[1].step(x, y);
````

For exponentially increasing data, it is sometimes more appropriate to plot on a log scale:

````python
x = np.array([1, 10, 100, 1000, 10000, 100000])
y = np.array([1.001, 10.1, 99.8, 999.6, 10004.5, 100008])
````

The Axes plotting method semilogx is equivalent to the plot method with the Axes ```set_xscale``` method set to log:

````python
fig, ax = plt.subplots(nrows=1, ncols=2);
ax[0].plot(x, y);
ax[0].set_xscale('log')
ax[1].semilogx(x, y);
````

The Axes plotting method semilogy is equivalent to the plot method with the Axes ```set_yscale``` method set to log:

````python
fig, ax = plt.subplots(nrows=1, ncols=2);
ax[0].plot(x, y);
ax[0].set_yscale('log')
ax[1].semilogy(x, y);
````

The Axes plotting method loglog is equivalent to the plot method with the Axes ```set_xscale``` and ```set_yscale``` methods set to log:

````python
fig, ax = plt.subplots(nrows=1, ncols=2);
ax[0].plot(x, y);
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[1].loglog(x, y);
````

## Collections

Returning to linear data:

````python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0.001, 2.001, 3.999, 5.999, 8.002, 10.001])
````

The Axes plotting method scatter can be used to create a scatter plot. The scatter plot using only the positional input arguments x and y is similar to a line plot with a line style of None and a marker:

````python
? ax.scatter
````

````python
fig, ax = plt.subplots();
points = ax.scatter(x, y);
````

Unlike plot, the scatter function has no multi-form input and returns a single collection of points:

````python
points
````

if the non-callable identifiers of the Axes are examined, there is a collections list:

````python
for identifier in dir(plt.Axes):
    isfunction = callable(getattr(plt.Axes, identifier))
    if not isfunction:
        print(identifier, end=' ')
        
    
````

The collection of points used for the scatter plot are included here:

````python
ax.collections
````

The collection of points can be accessed from index 0 of this list:

````python
ax.collections[0]
````

````python
fig, ax = plt.subplots();
ax.scatter(x, y);
ax.scatter(x, y+1);
````

Every time a collection of points is added to an Axes, they are appended to the collections list:

````python
ax.collections
````

The properties can be examined for this using:

````python
plt.getp(ax.collections)
````

The pyplot function set properties setp can be used to change the properties of all the traces:

````python
fig, ax = plt.subplots();
ax.scatter(x, y);
ax.scatter(x, y+1);
plt.setp(ax.collections, color='tomato');
````

Alternatively a single collection can be accessed by indexing into this list:

````python
plt.getp(ax.collections[0])
````

````python
fig, ax = plt.subplots();
ax.scatter(x, y);
ax.scatter(x, y+1);
plt.setp(ax.collections[0], color='tomato');
````

The Axes scatter methods keyword input argument s is equivalent to the Axes plot keyword input argument markersize but is more powerful and can be a scalar or a vector:

````python
fig, ax = plt.subplots();
ax.scatter(x, y);
ax.scatter(x, y+1, s=150);
ax.scatter(x, y+2, s=10*y+1);
````

The Axes scatter methods keyword input argument c is equivalent to the Axes plot keyword input argument markerfacecolor but is more powerful and can be a scalar or a vector of colors:

````python
from matplotlib.colors import CSS4_COLORS
np.random.seed(2)
colors = np.random.choice(list(CSS4_COLORS.keys()), size=x.shape)
colors
````

````python
fig, ax = plt.subplots();
ax.scatter(x, y, c='royalblue');
ax.scatter(x, y+1, c=colors);
````

The Axes scatter methods keyword input argument edgecolor is equivalent to the Axes plot keyword input argument markeredgecolor but once again is more powerful and can be a scalar or a vector of colors:

````python
edgecolors = np.random.choice(list(CSS4_COLORS.keys()), size=x.shape)
edgecolors
````

````python
fig, ax = plt.subplots();
ax.scatter(x, y, s=100, c='royalblue', edgecolor='tomato', linewidths=2);
ax.scatter(x, y+1, s=100, c=colors, edgecolors=edgecolors, linewidths=2);
````

The Axes scatter methods keyword input argument linewidths is equivalent to the Axes plot keyword input argument markeredgewidth but once again is more powerful and can be a scalar or a vector of colors. It should not be confused with the Axes plot keyword input argument linewidth which acts on the line and not the marker:

````python
fig, ax = plt.subplots();
ax.scatter(x, y, s=100, c='royalblue', edgecolor='tomato', linewidths=2);
ax.scatter(x, y+1, s=100, c=colors, edgecolors=edgecolors, linewidths=0.5*y);
````

The Axes scatter methods keyword input argument marker is equivalent to the Axes plot keyword input argument marker and takes the same values. In both cases only a scalar is accepted:

````python
fig, ax = plt.subplots();
ax.scatter(x, y, s=100, c='royalblue', edgecolor='tomato', linewidths=2, marker='s');
ax.scatter(x, y+1, s=100, c=colors, edgecolors=edgecolors, linewidths=0.5*y, marker='p');
````
